# Aprendizagem de Máquina

É uma linha de pesquisa da área de Inteligência Artificial que trata de ensinar modelos com dados de exemplos e melhorar a acurácia numa aplicação alvo.

## Aprendizado indutivo 
Dizemos que o aprendizado pode ser expresso por uma função com pares de entrada e saída.

Aqui existe o mapeamento de um conjunto (amostra) de variáveis de entrada de distribuição desconhecida para um conjunto de valores de saída.

**Objetivo:** Encontrar uma hipótese h que melhor se ajuste a função objetivo f (desconhecida) dentre as infinitas hipoteses no espaço de hipóteses H. Então, dada uma coleção de exemplos de f, retornar uma função h que melhor se aproxime de f.

Observe no gráfico abaixo que temos diferentes funções h do espaço H de hipóteses que se aproximam de f em graus diferentes. A pergunta que fica é **COMO escolher a função h que mais se aproxima de f?**

Perceba que as diferentes h's tem diferentes ajustes aos exemplos amostrados.

**Hipótese consistente:** É aquela que passa por todos os x's da amostra de entrada. No exemplo abaixo, seriam h3 e h4.

Porém, qual das duas deve ser escolhida se as duas são consistentes (que é o primeiro critério de avaliação de boas hipóteses)? 

**Lâmina de Ockham:** Após identificar as hipóteses consistentes devemos escolher a mais simples. Diz-se que hipóteses mais complexas estão mais ajustadas aos dados de amostra e tem menor poder de generalização.


**Por quê escolhemos hipóteses consistentes mais simples?**
- Então escolhemos hipóteses consistentes mais simples pelo simples fato de que ao treinar o modelo com um conjunto A de entrada, a acurácia pode não ser boa ao executar o modelo num conjunto B de treino caso ele seja muito diferente;
- A complexidade computacional da função reflete no tempo de execução do modelo para teste e treinamento, por isso, escolher funções menos complexas, como são as polinomiais é uma melhor escolha;

Mas **quem gera o espaço H?** A resposta é simples, as infinitas hipóteses h são geradas pelo algorítmo de aprendizagem utilizado, bem como pela busca de parâmetros.

**Problema Realizável:** Um problema de aprendizado é realizável se o espaço de Hipóteses contém a função objetivo. Então:

H = {h1, h2, h3,... ,hn}

f(x) = h3(x)

### Aprendizado supervisionado

Neste tipo de aprendizado nós não conhecemos nossa função objetivo, como já foi dito, mas dessa vez nós conhecemos nosso conjunto de saída Y. Podemos então apontar para o modelo qual resultado está certo e qual está errado.

**Saída Y Discreta:** Nossa saída y pode ser categórica, o que leva a uma tarefa de Classificação
**Saída Y Contínua:** Nossa saída y pode ser contínua, o que leva a uma tarefa de Regressão

### Aprendizado não supervisionado

Neste tipo de aprendizado nós não conhecemos nosso conjunto de saída Y. No entanto, conhecemos o comportamento do conjunto de entrada X. É através dos padrões de comportamento de X que nosso modelo aprende, sem qualquer tipo de feedback humano. O tipo mais comum de tarefa nesse caso é o de Clusterização (Agrupamento).

Existem três principais tarefas para um modelo não supervisionado:
- Agrupamento: Agrupar dados de acordo com características padrão
- Recuperação: Buscar um dado com base em características semelhantes às passadas como parâmetro de busca
- Regras de Associação: Se A tem isso, então B tem aquilo

### Aprendizado semi-supervisionado

Neste caso é nos dado um conjunto de entrada em que alguns poucos dados são conhecidos, enquanto que a maioria não. Então poucos deles tem informação de y e outros não. Pense no seguinte: Devemos construir um sistema que adivinhe a idade das pessoas pela foto. Nesse sistema, algumas imagens tem metadados com a idade e outras não. Se para coletar esses dados nós perguntamos aos entrevistados suas idades no momento da foto, isso é aprendizado supervisionado. Mas depois descobrimos que na verdade algumas pessoas mentiram suas idades. Não são apenas ruídos, é um problema sistemático e para resolvê-lo é um problema de aprendizagem não supervisionado envolvendo imagens, idades autodeclaradas e idades verdadeiras. 

### Aprendizado por reforço

Os agentes resolvedores de problema aprendem de uma série de reforços por - recompensa ou punição. Damos uma oportunidade de um modelo decidir qual ação levou a uma recompensa ou punição e estimulá-lo a continuar progredindo. É o aprendizado pela **experiência**. 

Nesse caso temos uma Função de Perda que recebe como entrada as informações sobre o efeito que uma ação causou no ambiente e calcula a distância para o objetivo final.

Em cada passo, o agente:
1. Executa ação;
2. Observa o efeito;
3. Recebe Recompensa/Punição.

### Árvores de Decisão

É simplesmente um dos algorítmos mais simples e bem sucedidos em aprendizagem de máquina.

Uma árvore de decisão representa uma função que recebe um vetor de atributos e retorna uma decisão como um único valor de saída, que pode ser contínuo ou discreto. 

**Ponto positivo**
Podemos pensar em um problema, cuja entrada da árvore é dada em valores discretos e a saída em valores booleanos.
Uma árvore de decisão alcança seu objetivo executando uma sequência de testes. 

**Nó interno:** Cada nó interno corresponde a um teste de um dos atributos de entrada, Ai.

**Arestas:** E as arestas são nomeadas com os possíveis valores do atributo Ai = Vik.

**Folhas:** Cada folha corresponde a um valor a ser retornado pela função.

**Predicado (o que se afirma do sujeito) de meta:** Como exemplo vamos construir uma árvore de decisão para decidir se esperamos por uma mesa no restaurante. O objetivo aqui é aprender uma definição para o predicado de meta WillWait. Primeiro vamos listar os atributos que consideramos uma parte da entrada:
1. Alternate: Se há algum restaurante adequado por perto
2. Bar: Se o bar possui um bar confortável para esperar por uma mesa
3. Fri/Sat: True na sexta e sábado
4. Hungry: Se estamos com fome
5. Patrons: Quantas pessoas estão no restaurante (valores podem ser None, Some e Full)
6. Price: A faixa de preço do restaurante ($, $$, $$$)
7. Raining: Se está chovendo lá fora
8. Reservation: Se fizemos uma reserva
9. Type: O tipo de restaurante (Francês, Italiano, Thailandesa ou Hamburgueria)
10. WaitEstimate: A espera estimada pelo anfitrião (0 - 10 minutes, 10-30, 30-60, or >60)

Note que cada atributo tem nos galhos (arestas) um pequeno conjunto de possíveis valores em vez de um valor inteiro.
Observe também que os atributos Price e Type foram ignorados. Os exemplos são processados pela árvore iniciando da raiz e seguindo o ramo (aresta) apropriado até que a folha seja alcançada.

Qualquer função booleana pode ser escrita como uma árvore de decisão. Como,

Goal <=> (Path1 v Path2 v ...),

onde cada Path é uma conjunção do tipo atributo-valor de testes requisitados para seguir o caminho. Algo como,

Goal <=> ((Atributo1 /\ Valor1) V (Atributo2 /\ Valor2) V ...).

Portanto, toda a expressão é equivalente a uma forma normal disjuntiva, ou seja, qualquer função na lógica proposicional pode ser expressa como uma árvore de decisão. 

Trivialmente existe uma árvore de decisão consistente para qualquer conjunto de treinamento com um caminho para cada exemplo.

**Devemos procurar por árvores de decisão mais compactas**, pois a *complexidade de tempo para percorrê-la pode ser grande(isso foi algo q eu disse kk)*. Sua topologia deve ser a que mais se aproxima da base de treino, mas sem se aprofundar muito, ela deve ser mais geral.

O **ideal** é escolher os atributos que mais separam as variáveis.

Se, no caso de uma árvore que tem saída booleana, todos os exemplos restantes forem positivos ou negativos, então nós terminamos com SIM ou NÃO.

Se não resta nenhum exemplo, nenhum exemplo desse tipo foi observado. Então retorna-se um valor padrão calculado a partir da classificação de maioria no pai do nó.

#### Qual é o melhor atributo?

A técnica gulosa usada na construção (aprendizado) da árvore de decisão é utilizada para aproximar a minimização de profundidade final da árvore. A ideia é escolher o atributo que vai o mais longe possível ao prover uma classificação exata dos exemplos. Um atributo perfeito que divide os exemplos em conjuntos, cada um no qual é positivo ou negativo (folhas).

Precisamos de uma medida formal de "bastante bom" ou "realmente útil" para utilizar o que o livro chama de função IMPORTANCE.

A solução é baseada na notação do ganho de informação, definida nos termos da entropia.

**Entropia:** Medir os atributos a partir da quantidade esperada de informações que ele pode fornecer. É a quantia fundamental na teoria da informação. A entropia é a *medida de incerteza* de uma variável aleatória. Adquirir informação corresponde a redução na entropia. 

*Exemplo1:* Uma moeda que tem apenas um valor - cara - é uma moeda que não possui incerteza nenhuma, logo sua entropia é zero. Uma moeda que é lançada tem probabilidade igual de cair cara ou coroa, 0 ou 1, e nós veremos em breve que isso conta como "1 bit" de entropia.

*Exemplo2:* Um dado justo de quatro lados tem 2 bits de entropia, pois precisa de 2 bits para descrever um dos lados de igual probabilidade. 

**Bits de Informação**
Nossa formula abaixo calcula a quantidade de bits de informação que um atributo ou variável possui. Podemos afirmar que nosso conjunto A divide o conunto de treinamento E em subconjuntos de E1 a Ev, onde A pode ter v valores distintos. pi - iesimo valor positivo. ni - iesimo valor negativo.

#### Avaliação de desempenho

Como sabemos que a árvore h gerada é h~f?

Dizemos que um algorítmo de aprendizagem é bom se produz hipóteses que fazem um bom trabalho de previsão de classificações de exemplos não vistos.

**Método de avaliação**
1. Coletar um número de exemplos que descreva o comportamento do Mundo real;
2. Dividi-lo em conjunto de treinamento e conjunto de teste;
3. Aplicar o algorítmo de aprendizagem ao conjunto de treinamento, gerando uma hipótese h;
4. Medir a porcentagem de exemplos no conjunto de teste que são corretamente classificados por h;
5. Repetir as etapas de 1 a 4 para diferentes tamanhos de conjuntos de treinamento e de teste de cada tamanho selecionados aleatoriamente, ou seja, experimentar h em diferentes conjuntos de teste

**Curva de aprendizado:** É o percentual de classificações corretas sobre o conjunto de teste como uma função do tamanho do conjunto de treinamento.

**Overfitting:** É o sobreajuste nos dados de treinamento baseado em coincidencias na previsão de resultados pela hipótese. O overtiffing ocorre principalmente quando temos hipóteses que tem alta variação (sensibilidade alta as variações dos valores dos dados) ou quando temos coincidências em previsões devido a falta de padrão real a ser encontrado no conjunto de exemplos de teste.

    Por exemplo: Temos um dataset de dados que possuem atributos como peso, cor e tempo de finalização. Queremos prever se um dado irá retornar 6 ou não. A resposta ideal seria Não. Mas como uma árvore de decisões sempre tentará encontrar um padrão nos dados de treinamento, ela irá se basear em coincidências. Imagine que dois dados azuis de 7 gramas sejam lançados ao mesmo tempo e que os dois tenham retornado 6. A hipótese irá pensar que isso faz sentido e para cada dado com essas características, a resposta será 6, sendo que os dados são justos e existe um erro na previsão por conta desse viés.
    
**Obs:** O overfitting também tem mais possibilidade de ocorrer quando temos menos exemplos de teste e mais atributos, e menos possibilidade de ocorrer quando temos mais exemplos de teste e menos atributos.

## Redes Neurais Artificiais

A unidade basica do cérebro são os neuronios, e eles tem conexões e sinais. um neuronio sozinho não serve pra nada, mas em rede eles são capazes de fazer trabalho. Então entram sinais elétricos pelos dentritos, e são transmitidos para outros neurônios pelo terminal do axônio quando um limite elétrico é atingido. 

Simulamos uma rede neural natural em uma máquina usando as redes neurais artificiais para aprender.

**Atributos:** Um neurônio de perceptron tem atributos (variáveis);

**Função de Ativação:** Traz a não linearidade para a rede neural, se não tivesse função de ativação o sinal iria passar sempre e como em redes neurais naturais existe um parâmetro que define a passagem de sinal, na artificial também deverá existir, no caso a função de ativação define esse parâmetro.

    **f>=0**=1 Se a função de ativação for maior igual a 0 o sinal passa
    **f<0**=0 Se a função de ativação for menor que 0 o sinal não passa

Observe abaixo uma rede neural feita a partir de vários perceptrons, chamada de multilayer perceptron.

**Topologia:** A topologia influencia na capacidade e performance da rede neural. E não necessariamente mais camadas ou menos camadas diz algo. Para saber a quantidade de neuronios na rede, fazemos a soma da quantidade de atributos da camada de entrada mais a quantidade de classes de saída e dividimos por dois.

### Ajuste de pesos
**Algorítmo de Backpropagation:** Uma rede neural artificial tem uma característica diferente de outros modelos de aprendizado que é a propagação de erros. Você submete dados, os parâmetros dos neurônios são calculados pelas funções de ativação que propagam os resultados na saída. O valor esperado é comparado com o valor produzido e o erro é calculado a partir daí. O erro então é utilizado por uma função de perda que é propagada de volta pra rede neural pra correção dos pesos.

**Gradiente da função de perda:** É um vetor que conteḿ as derivadas parciais da função de perda em relação a cada um dos pesos da rede neural. Em outras palavras é uma taxa de variação da função de perda com relação a cada peso da rede;

**Descida do Gradiente Estocástico:** É um algorítmo de otimização usado em conjunto com a retropropagação. Ele atualiza iterativamente os pesos da rede movendo-se na direção oposta do gradiente da função de perda, mas em vez de calcular o gradiente na função de perda para todo o conjunto de dados de treinaemnto, ele calcula o gradiente para uma amostra aleatória dos dados (mini-batches). Isso torna o processo de treinamento mais eficiente em termos computacionais e ajuda a evitar mínimos locais;

**Epoch:** É a passagem de todos os registros/dados de uma base pela rede. Num classificador normal o processo de treinamento termina nessa passagem única. Em uma rede neural artifical podemos ter várias Epochs durante o treinamento. Ou seja, uma rede neural requer muito mais processamento.

### Parada

O treinamento pode ser parado pelo número de Epochs, por um tempo máximo ou quando uma métrica de erro atingir um valor específico (como MSE, MAE e MAPE)

## Deep Learning com Redes Neurais Convolucionais

É uma subárea de Machine Learning que se concentra no uso das Redes Neurais Artificiais profundas para aprender dados. É uma área dentro das RNA's para reconhecer padrões em imagens, processamento de linguagem natural, entre outros.

O Deep Learning diferente das Redes Neurais Artificiais utiliza RNA's PROFUNDAS, pois usa redes com várias camadas ocultas entre a entrada e saída. Essas redes profundas tem a capacidade de aprender representações de dados em vários níveis de abstração, permitindo o processamento de informações complexas e extração de caracterísitcas relevantes de forma automática.

As redes neurais profundas implementam uma rede chamada de rede neural convolucional. Trouxeram uma revolução na capacidade das máquinas de reconhecer imagens. Ela é uma série de transformações na imagem para daí entrar na rede neural.

Nós fazemos essas transformações antes, pois os atributos de entrada da rede seriam pixel por pixel caso não houvessem as camadas convolucionais e de pooling, o que leva a uma carga de dados muito grande. 

A rede neural convolucional tem maior acurácia, pois em vez de trabalhar com pixels, ela busca identificar características das imagens. Ela aprende a identificar características.

**Filter Maps:** É a primeira camada onde temos a convolução, a criação de vários mapas de características da imagem.

**Pooling:** Reduz a dimensionalidade e destaca as características mais importantes da imagem

**Flattening:** Pega as matrizes de características do Pooling e transforma em vetores, que são os atributos de entrada da rede neural artificial. Em resumo, as entradas da rede são características destacadas das imagens.
